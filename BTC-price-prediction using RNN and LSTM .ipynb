```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
#reading csv file
data = pd.read_csv(r'C:\Users\Arunabh\New folder\btc-usd-latest.csv', date_parser = True)
data.tail()
```

```python
data['Open'].plot(figsize=(16,6))
```

```python
# training data
data_training = data[data['Date']< '2020-01-01'].copy()
data_training
```

```python
# testing data
data_test = data[data['Date']> '2020-01-01'].copy()
data_test
```

```python
training_data = data_training.drop(['Date', 'Adj Close'], axis = 1)
training_data.head()
```

```python
#feature scaling
scaler = MinMaxScaler()
training_data = scaler.fit_transform(training_data)
training_data
```

```python
X_train = []
Y_train = []
```

```python
training_data.shape[0]
```

```python
#creating a data structure with 60 timestamps and 1 output
for i in range(60, training_data.shape[0]):
    X_train.append(training_data[i-60:i])
    Y_train.append(training_data[i,0])
```

```python
X_train, Y_train = np.array(X_train), np.array(Y_train)

X_train.shape
```

```python
# building the RNN
# importing keras libraries and packages
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
#initialising the RNN and adding the first LSTM layer and some dropout regularisation
regressor = Sequential()
regressor.add(LSTM(units = 50, activation = 'relu', return_sequences = True, input_shape = (X_train.shape[1], 5)))
regressor.add(Dropout(0.2))
```

```python
# adding the second,third and fourth LSTM layers
regressor.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
regressor.add(Dropout(0.3))

regressor.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
regressor.add(Dropout(0.4))

regressor.add(LSTM(units = 120, activation = 'relu'))
regressor.add(Dropout(0.5))

regressor.add(Dense(units =1))
```

```python
regressor.summary()
```

```python
# compiling the RNN
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# fitting the RNN to the training set
history=regressor.fit(X_train, Y_train, epochs = 20, batch_size =50)
```

```python
loss = history.history['loss']
epochs = range(len(loss))
plt.figure()
plt.plot(epochs, loss, label='Training loss')
plt.title("Training Loss")
plt.legend()
plt.show()
```
